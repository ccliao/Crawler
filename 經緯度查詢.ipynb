```python
import requests
import json
import pandas as pd
import csv
from time import sleep
from random import randint

#Google API網址 https://developers.google.com/maps/documentation/geocoding/start#GeocodingRequests

def get_lat_lng(address):
    address = address.split('(')[0]
    url ='https://maps.googleapis.com/maps/api/geocode/json?address='+address
    req = requests.get(url)
    
    if json.loads(req.text)['status'] == 'OK':
        jd = json.loads(req.text)['results']
        lat = jd[0]['geometry']['location']['lat']
        lng = jd[0]['geometry']['location']['lng']
        address_en = jd[0]['formatted_address']
        print (lat,lng,address)
        lat_lng.append(''.join(str(lat)+' '+str(lng)))
        address_eng.append(address_en)        
        return lat_lng, address_eng
    else:
        print (address+'：Something Wrong')
        
if __name__ == '__main__':
    with open('lion_location.csv', 'r') as lion_location_file:
        lion_location = pd.read_csv(lion_location_file)
        lat, lng, address_en, lat_lng, address_eng = [], [], [], [], []
        for address in lion_location['住址']:
            get_lat_lng(address)
#             lat_lng.append(' '.join(str(lat)+str(lng)))
#             address_eng.append(address_en)
            sleep(randint(1,3))
            
    lion_location_file.close()
    lion_location['lat_lng']=lat_lng
    lion_location['address_eng']=address_eng
    lion_location
    lion_location.to_csv('lion_location_latlng.csv', index=False)
    print ('儲存完成')
```
